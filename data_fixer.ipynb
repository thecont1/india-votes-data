```python
import pandas as pd
import json

target_dataset = "2024Assembly-MH"

df = pd.read_csv(f"./results/{target_dataset}.csv")
js = json.load(open(f"./results/{target_dataset}.json"))

df
```

```python
c_list = df['constituency'].unique()
df['constituency_no'] = None

for n, c in enumerate(c_list):
    if js['constituencywise_results'][n]['voting_data']['constituency'] == c:
        df.loc[df['constituency'] == c, 'constituency_no'] = n+1

df = df[["election_year", "election_type", "election_state", "constituency", "constituency_no", "serial_no", "candidate", "party", "evm_votes", "postal_votes"]]
display(df)
```

```python
df.to_csv(f"{target_dataset}-new.csv", index=False)
```
